#### Ejercicio 09. ¿Son los precios promedios de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios y escribe una conclusión sobre ellos.

```python
# Calcular el precio promedio para Valdemorillo
valdemorillo_data = ds[ds['level5'] == 'Valdemorillo']
precio_promedio_valdemorillo = valdemorillo_data['price'].mean()

# Calcular el precio promedio para Galapagar
galapagar_data = ds[ds['level5'] == 'Galapagar']
precio_promedio_galapagar = galapagar_data['price'].mean()

# Imprimir los precios promedios
print(f"Precio promedio en Valdemorillo: {precio_promedio_valdemorillo:.2f} EUR")
print(f"Precio promedio en Galapagar: {precio_promedio_galapagar:.2f} EUR")

# Conclusión
if abs(precio_promedio_valdemorillo - precio_promedio_galapagar) < 10000:
    print("\nConclusión: Los precios promedios en Valdemorillo y Galapagar son similares.")
else:
    print("\nConclusión: Los precios promedios en Valdemorillo y Galapagar son significativamente diferentes.")
```

#### Ejercicio 10. ¿Son los promedios de precio por metro cuadrado (precio/m2) de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios de precio por metro cuadrado y escribe una conclusión sobre ellos.

Pista: Crea una nueva columna llamada `pps` (*price per square* o precio por metro cuadrado) y luego analiza los valores.

```python
# Calcular el precio por metro cuadrado (pps)
ds['pps'] = ds['price'] / ds['surface']

# Calcular el precio promedio por metro cuadrado para Valdemorillo
valdemorillo_data = ds[ds['level5'] == 'Valdemorillo']
precio_pps_valdemorillo = valdemorillo_data['pps'].mean()

# Calcular el precio promedio por metro cuadrado para Galapagar
galapagar_data = ds[ds['level5'] == 'Galapagar']
precio_pps_galapagar = galapagar_data['pps'].mean()

# Imprimir los precios promedio por metro cuadrado
print(f"Precio promedio por metro cuadrado en Valdemorillo: {precio_pps_valdemorillo:.2f} EUR/m2")
print(f"Precio promedio por metro cuadrado en Galapagar: {precio_pps_galapagar:.2f} EUR/m2")

# Conclusión sobre los precios promedio por metro cuadrado
if abs(precio_pps_valdemorillo - precio_pps_galapagar) < 100:  # Se usa un umbral de 100 EUR/m2 para considerar "similares"
    print("\nConclusión: Los precios promedio por metro cuadrado en Valdemorillo y Galapagar son similares.")
else:
    print("\nConclusión: Los precios promedio por metro cuadrado en Valdemorillo y Galapagar son significativamente diferentes.")
```

#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
# Normalizar los precios por población usando z-score y trazar histogramas
plt.figure(figsize=(12, 8))
for i, poblacion in enumerate(cinturon_sur_poblaciones):
    precios_poblacion = ds_cinturon_sur.loc[ds_cinturon_sur['level5'] == poblacion, 'price']
    precios_normalizados = (precios_poblacion - precios_poblacion.mean()) / precios_poblacion.std() # Normalización Z-Score
    plt.subplot(2, 2, i + 1) # Crear subplot para cada población
    plt.hist(precios_normalizados, bins=10, alpha=0.7, edgecolor='black')
    plt.title(f'Histograma de Precios Normalizados - {poblacion}')
    plt.xlabel('Precio Normalizado')
    plt.ylabel('Frecuencia')
    plt.grid(True)
plt.tight_layout() 
plt.show()



# Forma de la distribución: Después de la normalización, los histogramas permiten comparar la forma de la distribución de precios entre las poblaciones, independientemente de las diferencias en la escala de precios original. Se puede observar si la distribución es simétrica, asimétrica, o si presenta múltiples picos.
# Dispersión: La dispersión de los datos normalizados es comparable entre las poblaciones, ya que la desviación estándar es 1 en todos los casos. Esto facilita la comparación de la variabilidad relativa de los precios.
# Valores atípicos: La normalización ayuda a identificar valores atípicos relativos dentro de cada población. Los valores que se alejan significativamente de 0 en el eje x son aquellos que se desvían mucho del precio medio de su respectiva población.
# Comparación de poblaciones: Aunque los precios originales tenían escalas diferentes, la normalización permite observar si, en términos relativos, hay diferencias en la concentración de precios entre las poblaciones. Por ejemplo, se puede identificar si los precios en una población tienden a agruparse más cerca de la media que en otra.

#Conclusión:
#La normalización de los precios revela que, si bien las poblaciones pueden tener rangos de precios absolutos diferentes, 
#la distribución relativa de los precios dentro de cada población comparte ciertas características. 
#Esto facilita la comparación de la forma de la distribución y la identificación de valores atípicos relativos.
```

#### Ejercicio 19. ¿Qué puedes decir sobre el precio por metro cuadrado (precio/m2) entre los municipios de 'Getafe' y 'Alcorcón'? Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

Pista: Crea una nueva columna llamada `pps` (price per square en inglés) y luego analiza los valores

```python
# Calcular el precio por metro cuadrado (pps) para el subconjunto del cinturón sur
ds_cinturon_sur.loc[:, 'pps'] = ds_cinturon_sur['price'] / ds_cinturon_sur['surface'] # Usa .loc para evitar la advertencia

# Calcular el precio promedio por metro cuadrado para Getafe
getafe_data = ds_cinturon_sur[ds_cinturon_sur['level5'] == 'Getafe']
precio_pps_getafe = getafe_data['pps'].mean()

# Calcular el precio promedio por metro cuadrado para Alcorcón
alcorcon_data = ds_cinturon_sur[ds_cinturon_sur['level5'] == 'Alcorcón']
precio_pps_alcorcon = alcorcon_data['pps'].mean()

# Imprimir los precios promedio por metro cuadrado
print(f"Precio promedio por metro cuadrado en Getafe: {precio_pps_getafe:.2f} EUR/m2")
print(f"Precio promedio por metro cuadrado en Alcorcón: {precio_pps_alcorcon:.2f} EUR/m2")

# Conclusión sobre los precios promedio por metro cuadrado en Getafe y Alcorcón
if abs(precio_pps_getafe - precio_pps_alcorcon) < 100:  # Umbral de 100 EUR/m2
    print("\nConclusión: Los precios promedio por metro cuadrado en Getafe y Alcorcón son similares.")
else:
    print("\nConclusión: Los precios promedio por metro cuadrado en Getafe y Alcorcón son significativamente diferentes.")
```

#### Ejercicio 21. Realiza un trazado de las coordenadas (columnas latitud y longitud) del cinturón sur de Madrid por color de cada población (debes usar el subconjunto obtenido del Ejercicio 14) (★★★★)

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
from ipyleaflet import Map, basemaps

# Mapa centrado en Madrid
map = Map(center = (40.35, -3.8), zoom = 11.5, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
map

```

```python
# Crear un diccionario para almacenar las coordenadas por población
coordenadas_por_poblacion = {}
for poblacion in cinturon_sur_poblaciones:
    poblacion_data = ds_cinturon_sur[ds_cinturon_sur['level5'] == poblacion]
    # Convertir latitud y longitud a listas, manejar valores nulos
    latitudes = poblacion_data['latitude'].dropna().tolist()
    longitudes = poblacion_data['longitude'].dropna().tolist()
    coordenadas_por_poblacion[poblacion] = {'latitudes': latitudes, 'longitudes': longitudes}

# Imprimir el diccionario
print("\nDiccionario de coordenadas por población:")
print(coordenadas_por_poblacion)

# Crear el mapa de Madrid
map_madrid = Map(center=(40.35, -3.8), zoom=11,  basemap=basemaps.OpenStreetMap.Mapnik)

# Agregar marcadores al mapa
from ipyleaflet import Marker
for poblacion, coordenadas in coordenadas_por_poblacion.items():
    for lat, lon in zip(coordenadas['latitudes'], coordenadas['longitudes']):
        marker = Marker(location=(lat, lon), title=poblacion)
        map_madrid.add_layer(marker)

# Mostrar el mapa
map_madrid

```
